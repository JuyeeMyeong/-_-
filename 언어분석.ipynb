```python
from wordcloud import WordCloud, STOPWORDS

import numpy as np
from PIL import Image
```

```python
import matplotlib.pyplot as plt
import platform

from matplotlib import font_manager, rc

if platform.system() == 'Darwin':
    rc('font', family = 'AppleGothic')
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf'
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown System')

%matplotlib inline
```

```python
plt.figure(figsize=(8,8))
plt.imshow(squid, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import nltk
from konlpy.corpus import kobill
files_ko = kobill.fileids()
```

```python
doc_ko = kobill.open('1809890.txt').read()
```

```python
doc_ko
```

```python
from konlpy.tag import Okt; o = Okt()
tokens_ko = o.nouns(doc_ko)
tokens_ko
```

```python
ko = nltk.Text(tokens_ko, name='대한민국 국회 의안 제 1809890호')
```

```python
print(len(ko.tokens))
print(len(set(ko.tokens)))
ko.vocab()
```

```python
plt.figure(figsize=(12, 6))
ko.plot(50)
plt.show()
```

```python
stop_words = ['.', '(', ')', ',', "'", '%', '-', 'X', ').', 'x', '의', '자', '에', '안', '번', '호', '을', '이', '다', '만', '로', '가', '를']
ko = [each_word for each_word in ko if each_word not in stop_words]
ko
```

```python
ko = nltk.Text(tokens_ko, name='대한민국 국회 의안 제 1809890호')
plt.figure(figsize=(12, 6))
ko.plot(50)
plt.show()
```

```python
ko.count('초등학교')
```

```python
plt.figure(figsize=(12,6))
ko.dispersion_plot(['육아휴직', '초등학교', '공무원'])
```

```python
ko.concordance('초등학교')
```

```python
data = ko.vocab().most_common(150)

wordcloud = WordCloud(font_path='/Library/Fonts/AppleGothic.ttf', relative_scaling = 0.2, background_color='white').generate_from_frequencies(dict(data))
```

```python
plt.figure(figsize=(12,8))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
